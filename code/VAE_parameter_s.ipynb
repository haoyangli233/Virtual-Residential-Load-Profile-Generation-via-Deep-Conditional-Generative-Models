```python
import torch
import json
import logging
import network
import utils

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(device)
```

# Data Loading

```python
medoid_loader = utils.get_medoid_loaders(64)
m2s_loader = utils.get_m2s_loader(128)

medoid_config_file = 'Config/medoid_config.json'
m2s_config_file = 'Config/m2s_config.json'
```

# Medoid VAE

```python
# Configure logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Define hyperparameter space
m_VAE_hspace = {
    'encoder_hidden': [256],
    'encoder_lstm_size': [64],
    'encoder_lstm_layers': [1, 2, 3, 4],
    'latent_dim': [64, 128, 256],
    'decoder_hidden_size': [1024],
    'embed_size': [4, 5, 6],
    'time_size': [128],
    'decoder_lstm_size': [8, 16, 32],
    'decoder_lstm_layers': [4, 5, 6, 7],
}

# Example call
best_config, best_loss = utils.mVAE_hsearch(network.m_VAE, m_VAE_hspace, medoid_loader, device, 'medoid_vae', 50)

with open(medoid_config_file, 'w') as f:
    json.dump(best_config, f, indent=4)
```
